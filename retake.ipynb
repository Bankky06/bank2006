**Question 1** - **25  marks**


A construction site keeps daily records of concrete truck deliveries. Each record contains the truck ID and the delivered volume (in cubic meters).

You are given the following list of delivery records:

deliveries = [
    ("T01", 5.5),
    ("T02", 7.0),
    ("T01", 4.5),
    ("T03", 6.0),
    ("T02", 5.0),
]

**(a) Write a function to compute total volumes — 12 marks**

Create a function named total_volume(deliveries).

The function must:


*   take the list of deliveries as an input parameter.
*   create an empty dictionary to store total volumes for each truck.
*   loop through each tuple in the list.
*   add the delivered volume to the correct truck ID inside the dictionary.


The function must return a dictionary in the following format:

{

    "T01": total volume delivered by T01,

    "T02": total volume delivered by T02,

    "T03": total volume delivered by T03
}

Example of expected output (your program must calculate it):
{
    "T01": 10.0,

    "T02": 12.0,

    "T03": 6.0
}


```python
def total_volume(deliveries):
    
    totals = {}


    for truck_id, volume in deliveries:
        
        if truck_id in totals:
            totals[truck_id] += volume
        
        else:
            totals[truck_id] = volume

    return totals
deliveries = [
    ("T01", 5.5),
    ("T02", 7.0),
    ("T01", 4.5),
    ("T03", 6.0),
    ("T02", 5.0),
]

print(total_volume(deliveries))


```


**(b) Find the truck with the highest total volume — 8 marks**

After calling your total_volume() function:

Write code (outside the function) to find which truck delivered the largest amount of concrete in total.


*   Store the truck ID and its total volume.
*   Print the result using this exact format:


Top truck: T02 delivered 12.0 m3 (Your numbers should match the actual computed values.)

```python
# call the function
totals = total_volume(deliveries)

# find the truck with the highest total volume
top_truck = max(totals, key=totals.get)
top_volume = totals[top_truck]

# print the result in the required format
print(f"Top truck: {top_truck} delivered {top_volume} m3")
```

Example format:

A small dam monitors the daily water level (in meters) for 14 days. The data is stored in a Python list:

water_levels = [2.1, 2.3, 2.8, 3.0, 3.5, 4.2, 4.8, 5.1, 4.9, 4.0, 3.7, 3.2, 2.9, 2.4]


The safe operating range for the dam is:

Below 2.5 m → Too Low (Risk of insufficient water supply)

2.5 m to 4.5 m → Normal

Above 4.5 m → High (Flood Risk)


**TASK 1 - 8 marks**

1.   Print the water level on Day 1 (index 0)
2.   Print the water level on the last day using negative indexing.
3.   Print the water levels for days 5 to 8 using a loop with indexing.
4.   Print the maximum water level and the day it occurred (use indexing, not built-in functions like .index()).





```python
water_levels = [2.1, 2.3, 2.8, 3.0, 3.5, 4.2, 4.8, 5.1, 4.9, 4.0, 3.7, 3.2, 2.9, 2.4]

print("Day 1 water level:", water_levels[0])

print("Last day water level:", water_levels[-1])

print("Water levels from Day 5 to Day 8:")
for i in range(4, 8):   # indices 4 to 7 correspond to days 5 to 8
    print(f"Day {i + 1}: {water_levels[i]}")

max_level = water_levels[0]
max_day = 1

for i in range(len(water_levels)):
    if water_levels[i] > max_level:
        max_level = water_levels[i]
        max_day = i + 1

print(f"Maximum water level: {max_level} m on Day {max_day}")
```

**TASK 2 - 8 marks**

Write a loop that goes through all 14 days and prints the risk category for each day using the rules:

If level < 2.5 → "Too Low"

If 2.5 ≤ level ≤ 4.5 → "Normal"

If level > 4.5 → "High"


Your output format should be:

Day 1: 2.1 m → Too Low

Day 2: 2.3 m → Too Low

Day 3: 2.8 m → Normal

```python
water_levels = [2.1, 2.3, 2.8, 3.0, 3.5, 4.2, 4.8, 5.1, 4.9, 4.0, 3.7, 3.2, 2.9, 2.4]

for i in range(len(water_levels)):
    level = water_levels[i]

    if level < 2.5:
        risk = "Too Low"
    elif level <= 4.5:
        risk = "Normal"
    else:
        risk = "High"

    print(f"Day {i + 1}: {level} m → {risk}")
```

```python
water_levels = [2.1, 2.3, 2.8, 3.0, 3.5, 4.2, 4.8, 5.1, 4.9, 4.0, 3.7, 3.2, 2.9, 2.4]

too_low_days = 0
normal_days = 0
high_days = 0


for level in water_levels:
    if level < 2.5:
        too_low_days += 1
    elif level <= 4.5:
        normal_days += 1
    else:
        high_days += 1

print(f"Too Low days: {too_low_days}")
print(f"Normal days: {normal_days}")
print(f"High days: {high_days}")
```

**PART 2 - 9 marks**

The engineer wants to store the maximum allowed load for each sensor as a private attribute.
They also want a method to check whether the latest reading exceeds the limit.

Fix the class so that:

The overload check works correctly with the private attribute.

The max load can be read from outside the class using a method (for example, a getter).

The code produces exactly the expected output shown above

```python
class BridgeSensor:
    def __init__(self, sensor_id, readings, max_load):
        self.sensor_id = sensor_id
        self.readings = readings
        self.__max_load = max_load   
    def get_max_load(self):
        return self.__max_load

    def is_overloaded(self):
        latest = self.readings[-1]
        return latest > self.__max_load  

s = BridgeSensor("S1", [2.0, 2.3, 3.1], 3.0)

print("Max load for sensor:", s.get_max_load())
print("Overloaded?", s.is_overloaded())

```

**PART 3 - 8 marks**

The engineer wants a method that returns both:

the average reading

and a status string "OK" or "OVERLOAD" based on __max_load.


Rewrite the average_status method so that:

It has the correct parameter list.

It correctly calculates the average using self.readings.

It returns two values: (avg, status) instead of printing them.

```python
class BridgeSensor:
    def __init__(self, sensor_id, readings, max_load):
        self.sensor_id = sensor_id
        self.readings = readings
        self.__max_load = max_load

    def average_status(self):
        total = 0
        for r in self.readings:
            total += r
        avg = total / len(self.readings)

        if avg <= self.__max_load:
            status = "OK"
        else:
            status = "OVERLOAD"

        return avg, status  

s = BridgeSensor("S2", [2.5, 3.0, 3.5], 3.0)

avg_value, status_text = s.average_status()
print("For sensor", s.sensor_id, "=>", avg_value, status_text)

```
